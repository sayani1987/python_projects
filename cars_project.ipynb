#Find null values and fill with mean of that column

## Engine size column have 4 nulls, so filling with mean()

```python
car.fillna(car['EngineSize'].mean(), inplace=True)
```

```python
car['EngineSize'].isnull().sum()
```

##Cylinders column have 6 nulls, so filling with mean()

```python
car.fillna(car['Cylinders'].mean(), inplace=True)
```

```python
car['Cylinders'].isnull().sum()
```

##Horse Power column have 4 nulls, so filling with mean()

```python
car.fillna(car['Horsepower'].mean(), inplace=True)
```

```python
car['Horsepower'].isnull().sum()
```
